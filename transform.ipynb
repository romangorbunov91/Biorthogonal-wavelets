# Daub 5/3 and Daub 9/7 systems

```python
import numpy as np
import matplotlib.pyplot as plt
```

Импорт пользовательских функций.

```python
from Functions.daub_5_3 import daub_5_3
from Functions.daub_9_7 import daub_9_7
from Functions.coif_6 import coif_6
from Functions.wave_vect import wave_vect
```

```python
# Number of points.
N = 1024
# Transform level.
K = 6
```

- Vdir - analysis scaling signals ${V^K}$
- Wdir - analysis wavelets ${W^K}$
- Vrev - synthesis scaling signals ${\tilde{V}^K}$
- Wrev - synthesis wavelets ${\tilde{W}^K}$

## Daub 5/3 ${V_9^6}$, ${W_9^6}$, ${\tilde{V}_9^6}$, ${\tilde{W}_9^6}$

```python
index = 9
level = 6

Hplot = np.empty((4,), dtype=object)

[Vdir, Wdir, Vrev, Wrev] = wave_vect(N, K, wave_type=daub_5_3)

Hplot[0] = Vdir[level-1][index-1]
Hplot[1] = Wdir[level-1][index-1]
Hplot[2] = Vrev[level-1][index-1]
Hplot[3] = Wrev[level-1][index-1]

ax_titles = [
    'Daub 5/3 analysis scaling signal ${V_9^6}$',
    'Daub 5/3 analysis wavelet ${W_9^6}$',
    r'Daub 5/3 synthesis scaling signal ${\tilde{V}_9^6}$',
    r'Daub 5/3 synthesis wavelet ${\tilde{W}_9^6}$']

fig, axs = plt.subplots(2, 2, figsize=(12, 9))

for idx, ax in enumerate(axs.flat):
    ax.plot(Hplot[idx].T)
    ax.set_title(ax_titles[idx])

plt.savefig('readme_img/daub_5_3_signals.png', bbox_inches='tight')
plt.show()
```

## Daub 9/7 ${V_9^6}$, ${W_9^6}$, ${\tilde{V}_9^6}$, ${\tilde{W}_9^6}$

```python
index = 9
level = 6

Hplot = np.empty((4,), dtype=object)

[Vdir, Wdir, Vrev, Wrev] = wave_vect(N, K, wave_type=daub_9_7)

Hplot[0] = Vdir[level-1][index-1]
Hplot[1] = Wdir[level-1][index-1]
Hplot[2] = Vrev[level-1][index-1]
Hplot[3] = Wrev[level-1][index-1]

ax_titles = [
    'Daub 9/7 analysis scaling signal ${V_9^6}$',
    'Daub 9/7 analysis wavelet ${W_9^6}$',
    r'Daub 9/7 synthesis scaling signal ${\tilde{V}_9^6}$',
    r'Daub 9/7 synthesis wavelet ${\tilde{W}_9^6}$']

fig, axs = plt.subplots(2, 2, figsize=(12, 9))

for idx, ax in enumerate(axs.flat):
    ax.plot(Hplot[idx].T)
    ax.set_title(ax_titles[idx])

plt.savefig('readme_img/daub_9_7_signals.png', bbox_inches='tight')
plt.show()
```
